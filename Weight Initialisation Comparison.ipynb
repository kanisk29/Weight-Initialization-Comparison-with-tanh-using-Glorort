```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout,Dense
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons
from sklearn.preprocessing import MinMaxScaler
import warnings
warnings.filterwarnings("ignore")
```

```python
X,y = make_moons(noise = 0.06,n_samples = 500)
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
mmx = MinMaxScaler()
x_train = mmx.fit_transform(x_train)
x_test = mmx.transform(x_test)
```

```python
model2 = Sequential()
model2.add(Dense(10, activation="tanh", input_dim=2,kernel_initializer="glorot_uniform"))
model2.add(Dense(10, activation="tanh",kernel_initializer="glorot_uniform"))
model2.add(Dense(10, activation="tanh",kernel_initializer="glorot_uniform"))
model2.add(Dense(1, activation="sigmoid",kernel_initializer="glorot_uniform"))
model2.compile(optimizer='adam', metrics=['accuracy'],loss = 'binary_crossentropy')
history2 = model2.fit(x_train, y_train, epochs=100, validation_split=0.2)
```

```python
model3 = Sequential()
model3.add(Dense(10, activation="tanh", input_dim=2, kernel_initializer="glorot_normal"))
model3.add(Dense(10, activation="tanh", kernel_initializer="glorot_normal"))
model3.add(Dense(10, activation="tanh", kernel_initializer="glorot_normal"))
model3.add(Dense(1, activation="sigmoid",kernel_initializer="glorot_normal"))
model3.compile(optimizer='adam', metrics=['accuracy'],loss = 'binary_crossentropy')
history3 = model3.fit(x_train, y_train, epochs=100, validation_split=0.2)
```

```python
plt.plot(history2.history['val_accuracy'],label = 'Model 1 with Glorot Uniform')
plt.plot(history2.history['accuracy'],label = 'Model 1 with Glorot Uniform')

plt.plot(history3.history['val_accuracy'],label = 'Model 2 with Glorot Normal')
plt.plot(history3.history['accuracy'],label = 'Model 2 with Glorot Normal')

plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.plot(history2.history['val_loss'],label = 'Model 1 with Glorot Uniform')
plt.plot(history2.history['loss'],label = 'Model 1 with Glorot Uniform')

plt.plot(history3.history['val_loss'],label = 'Model 2 with Glorot Normal')
plt.plot(history3.history['loss'],label = 'Model 2 with Glorot Normal')

plt.grid(True)
plt.legend()
plt.show()
```
